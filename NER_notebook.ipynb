```python
# Imports & setup

import spacy
from tabulate import tabulate

from pprint import pprint
nlp = spacy.load('la_core_web_lg')
files= ['/content/drive/MyDrive/TXT/DP002_Tertapol.txt']
#files = ['extrait_C_II_Aretius/C_II_v_8.txt']

for filename in files:
    print(f"\n{'-'*50}")
    print(f"Processing: {filename}")

with open(filename, 'r', encoding='utf-8') as f:
    text = f.read()
    doc = nlp(text)
```

```python
from spacy import displacy
```

```python
# Extract entities

text =  text.replace("v","u").replace("V","U")

doc = nlp(text)

data = []
tokens = [item for item in doc]

for token in tokens:
    data.append([token.text, token.ent_type_])

    print(tabulate(data, headers=['Text', "Entity Type"]))
```

```python
# Save to CSV
import csv

with open("NER1-Tertapol.csv", mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Text", "Entity Type"])  # headers
    writer.writerows("/content/drive/MyDrive/data")

    print("NER1-Tertapol.csv")
```
