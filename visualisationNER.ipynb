# Visualisation


## setup 
1. charger le document : 
    * files= ['chemin vers le fichier']


```python
# Imports & setup

import spacy
from tabulate import tabulate

from pprint import pprint
nlp = spacy.load('la_core_web_lg')
files= ['txt/DP002_Tertapol.txt']
#files = ['extrait_C_II_Aretius/C_II_v_8.txt']

for filename in files:
    print(f"\n{'-'*50}")
    print(f"Processing: {filename}")
   
    
    with open(filename, 'r', encoding='utf-8') as f:
        text = f.read()
doc = nlp(text)
print(doc)
```

2. importer la fonction displacy qui permet les visualisations 

```python
from spacy import displacy
```

# visualiser les annotations : 

```python
sents = list(doc.sents)

data = []

for token in sents[1]:
    data.append([token.i, token.text, token.dep_, token.head.text])
    
print(tabulate(data, headers=['','Token', 'Dep', 'Head']))
```

1. visualiser les liens entre les mots 
* pertient pour un contenu non-lemmatisés, car représente la structure grammaticale de la phrase 

```python
displacy.render(sents[1], style='dep', jupyter=True)
```

2. visualiser les entitées nommées 

```python
# Extract entities

text =  text.replace("v","u").replace("V","U")

doc = nlp(text)

data = []
tokens = [item for item in doc]

for token in tokens:
    data.append([token.text, token.ent_type_])    

print(tabulate(data, headers=['Text', "Entity Type"])) 
```

Sauvegarder le résultat dans un csv 

```python
# Save to CSV
import csv

with open("NER-1-Tertapol.csv", mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Text", "Entity Type"])  # headers
    writer.writerows(data)

print("NER-1-Tertapol.csv")
```

```python
# Create filtered data (only rows with entities)
entity_data = [row for row in data if row[1] != ""]

# Save filtered CSV
with open("NER-1-Tertapol_entities.csv", mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Text", "Entity Type"])
    writer.writerows(entity_data)

print("NER-1-Tertapol_entities.csv")
print(tabulate(entity_data, headers=["Text", "Entity Type"]))
```
